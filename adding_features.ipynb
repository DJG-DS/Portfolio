Create a full lookup table for LTLA and UTLA codes.

```python
# import data and format to desired format
df_1 = pd.read_csv('D:/Users/daniel.godden/Data/data/Lower_Tier_Local_Authority_to_Upper_Tier_Local_Authority_(December_2022)_Lookup_in_England_and_Wales.csv')
df_1 = df_1.loc[:, ['LTLA22CD','LTLA22NM','UTLA22CD','UTLA22NM' ]]
df_1 = df_1.rename(columns={'LTLA22CD':'LTLA_Code'})
df_1 = df_1.rename(columns={'LTLA22NM':'LTLA'})
df_1 = df_1.rename(columns={'UTLA22CD':'UTLA_Code'})
df_1 = df_1.rename(columns={'UTLA22NM':'UTLA'})
df_1.head(5)
```

```python
# import data and format to desired format
df_2 = pd.read_csv('D:/Users/daniel.godden/Data/data/Local_Authority_Districts_(December_2022)_Names_and_Codes_in_the_United_Kingdom.csv')
df_2 = df_2.loc[:, ['LAD22CD','LAD22NM']]
df_2 = df_2.rename(columns={'LAD22CD':'LTLA_Code'})
df_2 = df_2.rename(columns={'LAD22NM':'LTLA'})
df_2.head(5)
```

```python
# merge dataframes into one
df = pd.merge(df_1.drop_duplicates(),df_2, on=['LTLA_Code', 'LTLA'], how='outer')
df.head(5)
```

```python
# import data and format to desired format
df_3 = pd.read_csv('D:/Users/daniel.godden/Data/data/Upper_Tier_Local_Authorities_(Dec_2022)_Names_and_Codes_in_the_United_Kingdom.csv')
df_3 = df_3.loc[:, ['UTLA22CD','UTLA22NM']]
df_3 = df_3.rename(columns={'UTLA22CD':'UTLA_Code'})
df_3 = df_3.rename(columns={'UTLA22NM':'UTLA'})
df_3.head(5)
```

```python
# merge dataframes into one
df = pd.merge(df,df_3.drop_duplicates(), on=['UTLA_Code', 'UTLA'], how='outer')
df.head(5)
```

```python
# populate the missing rows of LTLA/UTLA for Scotland/NI
df['LTLA_Code'] = df.apply(lambda row: row['UTLA_Code'] if pd.isnull(row['LTLA_Code']) else row['LTLA_Code'], axis=1)
df['UTLA_Code'] = df.apply(lambda row: row['LTLA_Code'] if pd.isnull(row['UTLA_Code']) else row['UTLA_Code'], axis=1)
df['LTLA'] = df.apply(lambda row: row['UTLA'] if pd.isnull(row['LTLA']) else row['LTLA'], axis=1)
df['UTLA'] = df.apply(lambda row: row['LTLA'] if pd.isnull(row['UTLA']) else row['UTLA'], axis=1)
```

```python
# create a 'Local_Authority_Code' column for merging into subnat data
df['Local_Authority_Code'] = df['LTLA_Code']

df.head(5)
```

Merge this dataframe with df_0 so that only required codes are included.

```python
# merge df into df_0
df0 = pd.merge(df_0,df.drop_duplicates(), on='Local_Authority_Code', how='left')
df0.head(5)
```

Include 'Fire&Rescue' to dataframe.

```python
# upload fire and rescure authority data
df_4 = pd.read_csv('D:/Users/daniel.godden/Data/data/Fire_and_Rescue_Authorities_(December_2022)_Names_and_Codes_in_the_United_Kingdom.csv')
df_4 = df_4.loc[:, ["FRA22CD","FRA22NM"]]
df_4 = df_4.rename(columns={"FRA22CD":"F&RA_Code"})
df_4 = df_4.rename(columns={"FRA22NM":"F&RA"})
df_4['Local_Authority_Code'] = df_4['F&RA_Code']
df_4.head(5)
```

```python
# merge df into df_000
df1 = pd.merge(df0,df_4.drop_duplicates(), on='Local_Authority_Code', how='left')
df1.head(5)
```

Merge the new features into the subnat_expenditure_data dataset and save as a csv file.

```python
# merge dataframes based on the Local_Authority_Code
df2 = pd.merge(data,df1.drop_duplicates(), on='Local_Authority_Code', how='inner')
df2.head(5)
```

```python
# specify the folder and file name to save the csv file
folder_path = 'D:/Users/daniel.godden/Data/output'
file_name = 'subnat_with_features.csv'

df2.to_csv(folder_path + '/' + file_name, index=False)
```
